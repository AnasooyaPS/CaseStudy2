# 3. Who scores the most on average for math, reading and writing based on 

# a)Gender

```python
df1=pd.DataFrame(data,columns=['gender','math score','reading score','writing score'])
```

```python
df1.groupby('gender').mean()
```

# b)Test preparation course

```python
df2=pd.DataFrame(data,columns=['test preparation course','math score','reading score','writing score'])
```

```python
df2.groupby('test preparation course').mean()
```

# 4. What do you think about the scoring variation for math, reading and writing based on

# a)Gender

```python
df1[['gender','math score','reading score','writing score']].groupby('gender').var()
```

# b)Test preparation course

```python
df2[['test preparation course','math score','reading score','writing score']].groupby('test preparation course').var()
```

# Inferences
* 482 Males and 518 Females participated in the test
* Regarding the "Parental level of education" ,the ones who gave the entry as 'Some college' had the most frequency(ie Mode of the dataset) and it is 226.This was followed by "Associate's degree" and "Highschool"
* For the averages for Math,Reading and writing based on gender,females fared better in reading and writing score with  averages around 72.59 and 72.47 respectively whereas the males fared better in math score with an average of around 68.82.
* For the averages for Math,Reading and writing based on test preparation course,those who had completed the test preparation course had better averages in all the three sections with averages of 69.97,74.18 and 74.68 in math,reading and writing respectively.
* Females had better scoring variance in math,reading and writing based on gender.
* Those who gave the test preparation entry as none had a better scoring variation in math,reading and writing based on test preparation.
